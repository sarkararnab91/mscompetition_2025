```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import f1_score, classification_report
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel("./TRAIN_NEW/combined_new.xlsx")

# Drop missing target values
df = df.dropna(subset=['Sex_F', 'ADHD_Outcome'])

# Define features and targets
X = df.drop(columns=['participant_id', 'Sex_F', 'ADHD_Outcome'], errors='ignore')
y = df[['Sex_F', 'ADHD_Outcome']]

# Encode categorical features
X_encoded = pd.get_dummies(X, drop_first=True)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Define sample weights
weights = np.where((y_train['Sex_F'] == 1) & (y_train['ADHD_Outcome'] == 1), 2, 1)

# Train MultiOutput model with RandomForest
base_model = RandomForestClassifier(n_estimators=100, random_state=42)
multi_model = MultiOutputClassifier(base_model)

# Fit model with sample weights (note: MultiOutputClassifier doesn't natively support weights, so we train manually)
multi_model.estimators_ = []
for i, col in enumerate(y_train.columns):
    model = base_model.__class__(**base_model.get_params())
    # Use adjusted weights only for ADHD, equal for Sex_F
    if col == 'ADHD_Outcome':
        model.fit(X_train, y_train[col], sample_weight=weights)
    else:
        model.fit(X_train, y_train[col])  # train without weighting
    multi_model.estimators_.append(model)

# Predict
y_pred = np.column_stack([model.predict(X_test) for model in multi_model.estimators_])
y_pred_df = pd.DataFrame(y_pred, columns=['Sex_F', 'ADHD_Outcome'])

# Weighted F1 Scores
f1_sex = f1_score(y_test['Sex_F'], y_pred_df['Sex_F'], average='weighted')
f1_adhd = f1_score(y_test['ADHD_Outcome'], y_pred_df['ADHD_Outcome'], average='weighted')
final_score = (f1_sex + f1_adhd) / 2

print("=== Classification Report for Sex_F ===")
print(classification_report(y_test['Sex_F'], y_pred_df['Sex_F']))

print("=== Classification Report for ADHD_Outcome ===")
print(classification_report(y_test['ADHD_Outcome'], y_pred_df['ADHD_Outcome']))

print(f"\nWeighted F1 Score (Sex_F): {f1_sex:.4f}")
print(f"Weighted F1 Score (ADHD): {f1_adhd:.4f}")
print(f"\nFinal Kaggle Leaderboard Score: {final_score:.4f}")


# Feature importance for SEX prediction
sex_model = multi_model.estimators_[0]
importances_sex = pd.Series(sex_model.feature_importances_, index=X_encoded.columns).sort_values(ascending=False)

# Plot top features
importances_sex.head(20).plot(kind='barh', title="Top Features for Predicting SEX")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()


# Feature importance for ADHD prediction
adhd_model = multi_model.estimators_[1]
importances = pd.Series(adhd_model.feature_importances_, index=X_encoded.columns).sort_values(ascending=False)

# Plot top features
importances.head(20).plot(kind='barh', title="Top Features for Predicting ADHD")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
